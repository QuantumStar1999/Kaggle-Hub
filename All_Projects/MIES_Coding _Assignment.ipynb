```python
# now To find point about which info_gain maximises

def find_max_gain_info(criteria,points):
  calc_info_gain=[]
  for i in points:
    info_gain=splitting(criteria,i)
    calc_info_gain.append(info_gain)
  plt.plot(points,calc_info_gain)
  plt.xlabel('points of '+criteria)
  plt.ylabel("information gain")
  max_gain=max(calc_info_gain)
  index=calc_info_gain.index(max_gain)
  return max_gain, points[index]
```

$${\displaystyle F_{1}={\frac {2}{\mathrm {recall^{-1}} +\mathrm {precision^{-1}} }}=2\cdot {\frac {\mathrm {precision} \cdot \mathrm {recall} }{\mathrm {precision} +\mathrm {recall} }}={\frac {\mathrm {tp} }{\mathrm {tp} +{\frac {1}{2}}(\mathrm {fp} +\mathrm {fn} )}}}$$


```python
def max_splitting2(criteria,points,mode=1):
  diff=[]
  f_score=[]
  for i in points:
    [_,_,pos]=splitting2(train_versicolor,criteria,i,mode)
    [_,_,neg]=splitting2(train_virginica,criteria,i,1-mode)
    if (pos+neg==0):
      diff.append(0)
      continue
    diff.append((pos-neg)/(pos+neg))
    fp=len(train_versicolor.index)-pos
    f_=pos/(pos+0.5*(fp+neg))
    f_score.append(f_)

  plt.plot(points,diff,'r--',points,f_score,'b.')
  plt.xlabel("points of "+criteria)
  plt.legend(['max split','f score'])
```

```python
class algebra:
  def __init__(self,x,y):
    self.v1=x
    self.v2=y
  mult=0
  def add(self):
    return(self.v1+self.v2)

  def mul(self,x,y):
    self.mult=(x*y)
    return self.mult

  def mul2(self):
    self.mult=(self.v1)*self.v2
    return self.mult
```

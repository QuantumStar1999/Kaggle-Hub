```python
def get_bagging_score(X,y,params):
    p1 = pd.DataFrame(extracted_models[0].predict_proba(X), )
    p2 = pd.DataFrame(extracted_models[1].predict_proba(X), )
    p3 = pd.DataFrame(extracted_models[2].predict_proba(X), )
    p4 = pd.DataFrame(extracted_models[3].predict_proba(X), )
    p5 = pd.DataFrame(extracted_models[4].predict_proba(X), )
    t = pd.get_dummies(y)
    a = params['a']
    b = params['b']
    c = params['c']
    d = params['d']
    e = params['e']
    f = (a+b+c+d+e)
    return roc_auc_score(t, a/f*p1+b/f*p2+c/f*p3+d/f*p4+e/f*p5, multi_class='ovr')
```

```python
def get_prediction(X,params):
    p1 = pd.DataFrame(extracted_models[0].predict_proba(X), )
    p2 = pd.DataFrame(extracted_models[1].predict_proba(X), )
    p3 = pd.DataFrame(extracted_models[2].predict_proba(X), )
    p4 = pd.DataFrame(extracted_models[3].predict_proba(X), )
    p5 = pd.DataFrame(extracted_models[4].predict_proba(X), )
    a = params['a']
    b = params['b']
    c = params['c']
    d = params['d']
    e = params['e']
    f = (a+b+c+d+e)
    return a/f*p1+b/f*p2+c/f*p3+d/f*p4+e/f*p5
```
